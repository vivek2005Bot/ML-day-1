# Correlation with Heatmap

A correlation heatmap uses colores cells,typically in a monochromatic scale, to show a 2D correlation matrix(table) between two discrete or event types.it is very important in Feature Selection.

```python
df_numeric = df.select_dtypes(include=[float, int])
df1= df_numeric.corr()
```

```python
df1.corr()
```

```python
sn.heatmap(df1.corr(),annot=True,fmt='.2f')
```

# Plotly

```python
import plotly.graph_objects as go
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3,4,5],y=[3,4,5,6,7],mode='markers'))
fig.show()
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3,4,5],y=[3,4,5,6,7],mode='lines'))
fig.show()
```

```python
fig=go.Figure()
fig.add_trace(go.Bar(x=[1,2,3,4,5],y=[3,4,5,6,7]))
fig.show()
```

```python
x=[1,2,3,4,5,6,8,5,67,23,59,0,34,6,9,5,2,44,9,10,12,15]
```

```python
fig=go.Figure(data=[go.Histogram(x=x)])
fig.show()
```

```python
fig=go.Figure(data=[go.Histogram(x=tips.total_bill)])
fig.show()
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill,y=tips.tip,mode='markers'))
fig.show()
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill,y=tips.tip,mode='markers',marker_size=10*tips['size']))
fig.show()
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter3d(x=tips.total_bill,y=tips.tip,mode='markers',z=tips['size']))
fig.show()
```
